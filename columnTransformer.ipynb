```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv(r"C:\Users\KIIT0001\Downloads\covid_toy.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
X_train
```

```python
#adding simple imputer to fever col
from sklearn.impute import SimpleImputer

# create imputer (default = mean)
si = SimpleImputer(strategy='mean')

# fit ONLY on training data
X_train_fever = si.fit_transform(X_train[['fever']])

# transform test data using same imputer
X_test_fever = si.fit_transform(X_test[['fever']])

# check shape
X_train_fever.shape
```

```python
#ordinalecoding -> cough
from sklearn.preprocessing import OrdinalEncoder

# define ordinal encoder with order
oe = OrdinalEncoder(categories=[['Mild', 'Strong']])

# fit ONLY on training data
X_train_cough = oe.fit_transform(X_train[['cough']])

# transform test data using same encoder
X_test_cough = oe.transform(X_test[['cough']])

# check shape
X_train_cough.shape
```

```python
# onehotencoding -> gener,city
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# create encoder
ohe = OneHotEncoder(drop='first', sparse_output=False)

# fit ONLY on training data
X_train_gender_city = ohe.fit_transform(X_train[['gender', 'city']])

# transform test data
X_test_gender_city = ohe.transform(X_test[['gender', 'city']])

# check shape
X_train_gender_city.shape
```

```python
#extracting Age
X_train_age=X_train.drop(columns=['gender','fever','cough','city']).values

X_test_age=X_test.drop(columns=['gender','fever','cough','city']).values
X_train_age.shape
```

```python
import numpy as np

# combine TRAIN features
X_train_transformed = np.concatenate(
    (X_train_age, X_train_fever, X_train_gender_city, X_train_cough),
    axis=1
)

# combine TEST features
X_test_transformed = np.concatenate(
    (X_test_age, X_test_fever, X_test_gender_city, X_test_cough),
    axis=1
)

# check shapes
X_train_transformed.shape, X_test_transformed.shape
```

```python
from sklearn.compose import ColumnTransformer
```
