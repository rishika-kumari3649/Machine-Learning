```python
import requests
```

```python
url = "https://openlibrary.org/search.json?q=python"  
```

```python
response = requests.get(url)
data = response.json()
```

```python
for i in range(5):
    print(data["docs"][i]["title"])
```

```python
import pandas as pd
```

```python

titles = [data["docs"][i]["title"] for i in range(10)]   # 10 titles
```

```python

df = pd.DataFrame({"title": titles})
df.to_csv("books.csv", index=False)
```
